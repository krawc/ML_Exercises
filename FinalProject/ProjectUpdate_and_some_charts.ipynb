```python
#BAR CHART PLOTTING FUNCTION

def plot_bar_from_counter(counter, title, number, barColor="r"):

    #add a subplot with a given number
    ax = fig.add_subplot(int(number))
    
    #get keys, values and x-axis ticks
    frequencies = counter.values()
    names = list(counter.keys())
    x_coordinates = np.arange(len(counter))

    #draw the thing
    ax.bar(x_coordinates, frequencies, align='center', alpha=0.7, color=barColor)
    ax.set_title(title, fontsize=24)
    ax.xaxis.set_major_locator(plt.FixedLocator(x_coordinates))
    ax.xaxis.set_major_formatter(plt.FixedFormatter(names))

    return ax
```

```python
#initialize counters
#data for months of the year
countMonthDel = Counter()
countMonthNon = Counter()
countMonthRatio = Counter()

#data for days of the week
countWeekDDel = Counter()
countWeekDNon = Counter()
countWeekDRatio = Counter()

#data for hours of the day
countHourDel = Counter()
countHourNon = Counter()
countHourRatio = Counter()

#data for airports
countPortDel = Counter()
countPortNon = Counter()
countPortRatio = Counter()

#data for airlines
countLineDel = Counter()
countLineNon = Counter()
countLineRatio = Counter()

#data for normalization
countflightspermonth = Counter()
countflightsperweek = Counter()
countflightsperhour = Counter()
countflightsperairl = Counter()
countflightsperairp = Counter()

#open the dataset (one with 12k examples, around 4k delayed and 8k non-delayed)
f1 = open('./12kdataset.csv', 'rt')

#read the data
dataPlot = csv.reader(f1)

for line in dataPlot:
    
    
    #convert hour:minute to just hour
    line[9] = math.floor(int(line[9])/100)
    
    # total flights
    countflightspermonth[line[1]] += 1
    countflightsperweek[line[3]] += 1
    countflightsperhour[line[9]] += 1
    countflightsperairl[line[4]] += 1
    countflightsperairp[line[7]] += 1
    countflightsperairp[line[8]] += 1
    
    
    #If a flight is delayed (the amount of delay > 0 minutes)
    if int(line[22]) > 0:
        
        #add flight data to counters
        countMonthDel[line[1]] += 1
        countWeekDDel[line[3]] += 1
        countLineDel[line[4]] += 1
        countPortDel[line[7]] += 1
        countPortDel[line[8]] += 1
        countHourDel[line[9]] += 1
    
    #If a flight starts on time, or early (the amount of delay <= 0 minutes)
    elif int(line[22]) <= 0:
        
        #add flight data to counters
        countMonthNon[line[1]] += 1
        countWeekDNon[line[3]] += 1
        countLineNon[line[4]] += 1
        countPortNon[line[7]] += 1
        countPortNon[line[8]] += 1
        countHourNon[line[9]] += 1
        
        #calculate delayed:nondelayed ratios
        countMonthRatio[line[1]] = countMonthDel[line[1]] / countMonthNon[line[1]]
        countWeekDRatio[line[3]] = countWeekDDel[line[3]] / countWeekDNon[line[3]]
        countLineRatio[line[4]] = countLineDel[line[4]] / countLineNon[line[4]]
        countPortRatio[line[7]] = countPortDel[line[7]] / countPortNon[line[7]]
        countPortRatio[line[8]] = countPortDel[line[8]] / countPortNon[line[8]]
        countHourRatio[line[9]] = countHourDel[line[9]] / countHourNon[line[9]]

f1.close()



#Sort across keys OR values depending on the desired data visualization

countMonthInt = {int(k): v for k, v in countMonthRatio.items()}
countWeekDInt = sorted(countWeekDRatio.items(), key=operator.itemgetter(0))
countHourInt = sorted(countHourRatio.items(), key=operator.itemgetter(0))
countPortSorted = sorted(countPortRatio.items(), key=operator.itemgetter(1), reverse=True)[:20]
countLineSorted = sorted(countLineRatio.items(), key=operator.itemgetter(1), reverse=True)

#Convert data to dictionaries to make them pluggable in matplotlib

countMonthDict = dict(countMonthInt)
countWeekDDict = dict(countWeekDInt)
countHourDict = dict(countHourInt)
countPortDict = dict(countPortSorted)
countLineDict = dict(countLineSorted)

#plot the figure
fig = plt.figure(figsize=(18.5, 27))

#create subplots
plot_bar_from_counter(countMonthDict, 'Delayed to non-delayed ratio per month of the year', 511, "#99e0c8")
plot_bar_from_counter(countWeekDDict, 'Delayed to non-delayed ratio per day of the week', 512, "#a3d4e6")
plot_bar_from_counter(countHourDict, 'Delayed to non-delayed ratio per hour of the day', 513, "#a768ff")
plot_bar_from_counter(countPortDict, 'Delayed to non-delayed ratio per airport', 514, '#a9b4f7')
plot_bar_from_counter(countLineDict, 'Delayed to non-delayed ratio per airline', 515, "#f0acbf")

plt.show()
```

### Preliminary data analysis & hypotheses:

1. We have two seasonal spikes in delayed flights. The first one is during the winter season - possibly because of the weather conditions and winter holidays. The second one is during the summer, especially in June - when many people travel for summer vacation.

2. For days of the week, Friday and Monday have the highest ratio of delayed to nondelayed flights, probably because the traffic is significantly higher right before and after the weekend.

3. Evening and late night flights are usually most likely to be delayed, either because of the increased traffic or other factors, such as low visibility, difficult weather conditions or workforce fatigue.

4. A majority of airports with a high number of delayed flights are regional, low-scale airports. It might be useful to check for inverse correlation between airport traffic and ratio of delayed to non-delayed flights.

5. Low-fare airlines tend to have a significantly greater amount of delays. The top three: Spirit Airlines (NK), Frontier (F9) and JetBlue (B6) are all low-cost operators that are more likely to compromise on customer experience to cut down maintenance costs. The biggest 3 airlines in the US - American, United and Delta are all less likely to delay, with Delta having the best score in this category.
